# **Configuração** **da** **planilha**

```python
import pandas as pd

nome_do_arquivo = "Produto.csv"

df = pd.read_csv(nome_do_arquivo, decimal=',')

df['Preço'] = df['Preço'].str.replace(',', '.', regex=True)
df['Preço'] = df['Preço'].str.replace('R\$', '', regex=True)
df['Preço'] = df['Preço'].str.replace('[^\d.]', '', regex=True)
df['Preço'] = pd.to_numeric(df['Preço'], errors='coerce')

print(df['Preço'])
```

# **Aplicação das Formulas**

```python
media = df['Preço'].mean()
desvio_padrao = df['Preço'].std()
```

```python
print("Média geral", media)
print(f"Desvio padrão geral", desvio_padrao)
```

```python
mediana = df['Preço'].median()
print("Mediana geral", mediana)
```

```python
Máximo = df['Preço'].max()
print("Máximo geral", Máximo)
minimo = df['Preço'].min()
print("Mínimo geral", minimo)
```

```python
amplitude = Máximo - minimo
print("Amplitude geral", amplitude)
```

# **Gráfico da Planilha**

```python
grouped = df.groupby('Marca')['Preço']
grouped.agg(['mean', 'std'])
```

```python
result = grouped.agg(['mean', 'std'])
result.columns = ['Média', 'Desvio Padrão']
```

```python
result.plot.bar()
```
